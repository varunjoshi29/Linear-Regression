### 1. (i) KNN Regression:  Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k  { 1 , 2 , . . . , 100 } that gives you the best fit. Plot the train and test errors in terms of 1/k .

```python
x_train, x_test, y_train, y_test = train_test_split(df[['AT','V','AP','RH']], df['PE'], test_size=0.3, random_state=42)
```

```python
err_best = float('inf')
k_best = None
train_errors = []
test_errors = []
k_to_test = [i for i in range(1,101)]
k_to_plot = []

for k in k_to_test:
    k_to_plot.append(1/k)
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(x_train, y_train)
    pred_train = model.predict(x_train)
    train_error = mean_squared_error(y_train, pred_train)
    train_errors.append(train_error)
    
    pred_test = model.predict(x_test)
    test_error = mean_squared_error(y_test, pred_test)
    test_errors.append(test_error)
    
    if test_error < err_best:
        err_best = test_error
        k_best = k

print(f"On Unnormalized Data, Best MSE achieved is {err_best} and obtained on k = {k_best} ")
```

```python
plt.figure(figsize=(10,6))
plt.plot(k_to_plot,test_errors, color='blue', label='Test Error')
plt.plot(k_to_plot,train_errors, color='red', label='Train Error')
plt.legend()
plt.title('Unnormalized Data: Error Rate vs. 1/K Value')
plt.xlabel('1/K')
plt.ylabel('Error Rate')
plt.show()
```

<b> Now, we will run the KNN regressor on normalized data </b>

```python
df_normalized = (df - df.mean())/ df.std()
x_train, x_test, y_train, y_test = train_test_split(df_normalized[['AT','V','AP','RH']], df['PE'], test_size=0.3, random_state=42)
```

```python
err_best = float('inf')
k_best = None
train_errors = []
test_errors = []
k_to_test = [i for i in range(1,101)]
k_to_plot = []

for k in k_to_test:
    k_to_plot.append(1/k)
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(x_train, y_train)
    pred_train = model.predict(x_train)
    train_error = mean_squared_error(y_train, pred_train)
    train_errors.append(train_error)
    
    pred_test = model.predict(x_test)
    test_error = mean_squared_error(y_test, pred_test)
    test_errors.append(test_error)
    
    if test_error < err_best:
        err_best = test_error
        k_best = k

print(f"On Normalized Data, Best MSE achieved is {err_best} and obtained on k = {k_best} ")
```

```python
plt.figure(figsize=(10,6))
plt.plot(k_to_plot,test_errors, color='blue', label='Test Error')
plt.plot(k_to_plot,train_errors, color='red', label='Train Error')
plt.legend()
plt.title('Normalized Data: Error Rate vs. 1/K Value')
plt.xlabel('1/K')
plt.ylabel('Error Rate')
plt.show()
```

### 2.  ISLR: 2.4.1

**(a) The sample size n is extremely large, and the number of predictors p is small. <br>**

**ANS: With large number of data points and a small number of predictors, it is better to use a flexible model as it will prevent overfitting that can cause in case of an inflexible model. This model will also have a better bias-variance trade off (with small p).**

**(b) The number of predictors p is extremely large, and the number of observations n is small. <br>**

**ANS: Flexible model will also fail on test data (will not have low test error and will overfit), but will perform better than inflexible model.**

**(c) The relationship between the predictors and response is highly non-linear.**

**ANS: Flexible model will be able to capture the complex non linear relations. There can be high variance in data in such case and inflexible model will have a huge bias and will underfit the data. Therefore Flexible model is better** 

**(d) The variance of the error terms, i.e. 2 = Var(), is extremely high.**

**ANS: A flexible model will also learn noise in data and therefore, will be too wiggly. It will not be able to generalize well and can perform poorly on test data.**
